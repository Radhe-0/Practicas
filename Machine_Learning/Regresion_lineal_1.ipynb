### Importo los datos y los guardo en un data frame

```python
from Generadores import *
import pandas as pd

# Cargo los datos y los coloco en un Data Frame

datos = DatosLineales(300)

df = pd.DataFrame(datos, columns=['X','Y'])

df.head()
```

### Grafico los datos para asegurarme de que siguen una cierta correlación lineal

```python
df.plot.scatter(x='X',y='Y',color='orange',edgecolors='indigo')
```

### Transformo los datos para poderlos usar en el modelo

```python
X = df['X'].to_numpy().reshape(-1,1)
Y = df['Y'].to_numpy()

from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X,Y)
predicciones = modelo.predict(X)

modelo.__dict__
```

$$\text{y = 7.18 + 4.89 x}$$

### Grafico la recta de regresión

```python
import matplotlib.pyplot as plt

plt.scatter(X,Y,color='orange',edgecolors='indigo')
plt.plot(X,predicciones,color='brown',linewidth=3)
plt.title('$y = 7.18 + 4.89 x$')
plt.grid()
plt.show()
```

### Precisión del modelo

```python
modelo.score(X,Y)
```

$\text{La precisión es del 88%}$
