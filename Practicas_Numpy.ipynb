```python
# Importo las librerías que usaré
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generador para Datos lineales

def DatosLineales(num=10):

    x = 2 * np.random.rand(num,1)
    x_ = [float(i) for i in x]
    y = np.random.randint(1,10) + np.random.randint(4,10) * x + np.random.randn(num,1)
    y_ = [float(i) for i in y]
    lista = []
    
    for i in zip(x_,y_):
        lista.append(i)

    return lista

datos = DatosLineales(350)
df = pd.DataFrame(datos, columns=['x','y'])
df
```

```python
df.plot.scatter(x='x',y='y',c='mediumturquoise',edgecolors='darkcyan',marker='D',linewidths=1.3)
```

```python
# Generador de datos no lineales

def DatosNoLineales(num=10):

    x = 2 * np.random.rand(num,1)
    x_ = [float(i) for i in x]
    y = x + 2 * x ** 3 + np.random.randn(num,1)
    y_ = [float(i) for i in y]
    lista = []
    
    for i in zip(x_,y_):
        lista.append(i)

    return lista

datos = DatosNoLineales(200)
df = pd.DataFrame(datos, columns=['x','y'])
df
```

```python
df.plot.scatter(x='x',y='y',c='khaki',edgecolors='olive',marker='s',linewidths=1.3)
```

```python
# Generadores de datos para practicar clasificación

def DatosClusters(num=10):

    numeros = np.random.rand(num)
    int_x = np.random.randint(0,100,size=num)
    int_y = np.random.randint(0,100,size=num)
    eje_x = np.sum([int_x, numeros], axis=0)
    eje_y = np.sum([int_y, numeros], axis=0)
    puntos = []

    for i in zip(eje_x, eje_y):
        puntos.append(i)
    
    return puntos

datos = DatosClusters(500)
df = pd.DataFrame(datos, columns=['x','y'])
df
```

```python
df.plot.scatter(x='x',y='y',c='tomato',edgecolors='maroon',marker='v',linewidths=1.3)
```

```python
# Es lo mismo que el anterior pero en 3D
def DatosClusters3D(num=10):
    
    lista = []
    
    numeros = np.random.rand(num)
    
    int_x = np.random.randint(0,100,size=num)
    int_y = np.random.randint(0,100,size=num)
    int_z = np.random.randint(0,100,size=num)
    
    eje_x = np.sum([int_x, numeros], axis=0)
    eje_y = np.sum([int_y, numeros], axis=0)
    eje_z = np.sum([int_z, numeros], axis=0)

    for i in zip(eje_x, eje_y, eje_z):
        lista.append(i)
    
    return lista

datos = DatosClusters3D(400)
df = pd.DataFrame(datos, columns=['X','Y','Z'])
df
```

```python
X = df['X']
Y = df['Y']
Z = df['Z']

ax = plt.axes(projection='3d')
ax.scatter(X,Y,Z, c='deepskyblue',edgecolors='black')
```

```python
# Este generador es para clasificadores del tipo Logistic Regression
def DatosLRegression(num=10):

    ceros = np.zeros(num)
    random_1 = np.random.rand(num)
    random_2 = np.random.randint(1,6,size=(1,num))
    valores = random_2 + random_1
    lista = []

    for i in zip(valores[0], ceros):
        lista.append(i)

    unos = np.ones(num)
    random_3 = np.random.rand(num)
    random_4 = np.random.randint(6,11,size=(1,num))
    valores_2 = random_4 + random_3

    for i in zip(valores_2[0], unos):
        lista.append(i)
    
    return lista

datos = DatosLRegression(20)
df = pd.DataFrame(datos, columns=['x','y'])
df.head(10)
```

```python
# Más datos para clasificadores
def DatosSVM(num=10):

    X = (np.r_[np.random.randn(num, 2) - [2, 2], np.random.randn(num, 2) + [2, 2]]).reshape(-1)
    Y = (np.r_[np.random.randn(num, 2) - [2, 2], np.random.randn(num, 2) + [2, 2]]).reshape(-1)
    lista = []

    for i in zip(X,Y):
        lista.append(i)

    return lista

datos = DatosSVM(150)
df = pd.DataFrame(datos, columns=['x','y'])
df
```

```python
df.plot.scatter(x='x',y='y',c='turquoise',edgecolors='darkslategrey',marker='^',linewidths=1.3)
```
