```python
from Generadores import *
import pandas as pd

# Cargo los datos y los coloco en un Data Frame

datos = DatosClusters(400)

df = pd.DataFrame(datos, columns=['X','Y'])

df.head()
```

## Visualizo los datos

```python
import matplotlib.pyplot as plt

df.plot.scatter(x='X',y='Y',color='orange',edgecolors='indigo')
```

## Creo el modelo y lo visualizo

```python
X = df.to_numpy()

from sklearn.cluster import KMeans

modelo = KMeans(n_clusters=6)
modelo.fit(X)
predicciones = modelo.predict(X)

import seaborn as sns

sns.scatterplot(x='X',y='Y',data=df,hue=predicciones,palette='tab10',marker='o')
plt.show()
```
