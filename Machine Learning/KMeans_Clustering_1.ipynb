```python
from Generadores import *
import pandas as pd

# Cargo los datos y los coloco en un Data Frame

datos = DatosClusters(400)

df = pd.DataFrame(datos, columns=['X','Y'])

df.head()
```

## Visualizo los datos

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x='X',y='Y',data=df,color='orange',edgecolors='indigo')
plt.grid()
plt.title('Datos')
plt.minorticks_on()
plt.plot()
```

## Creo el modelo y lo visualizo

```python
X = df.to_numpy()

from sklearn.cluster import KMeans

modelo = KMeans(n_clusters=8)
modelo.fit(X)
predicciones = modelo.predict(X)


colors = {0:'tab:blue', 
          1:'tab:orange', 
          2:'tab:green', 
          3:'tab:red', 
          4:'tab:purple', 
          5:'tab:brown', 
          6:'tab:pink',
          7: 'teal'}

fig, ax = plt.subplots(figsize=(12, 8))
df['color'] = predicciones

ax.scatter(df['X'],df['Y'], c=df['color'].map(colors),edgecolors='black')
plt.grid()
plt.minorticks_on()
plt.plot()
```
