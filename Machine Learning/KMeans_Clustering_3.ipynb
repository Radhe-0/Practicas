```python
import pandas as pd
from Generadores import *
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
datos = DatosSVM2(400)

df = pd.DataFrame(datos, columns=['X','Y'])

fig = plt.figure(figsize=(12,7))
ax = plt.subplot(1,1,1)
plt.scatter(df['X'],df['Y'],edgecolors='black')
plt.show()
```

```python
X = df.to_numpy()

modelo = KMeans(n_clusters=4)
modelo.fit(X)
pred = modelo.predict(X)
```

```python
colores = {
    0: 'crimson',
    1: 'orchid',
    2: 'mediumpurple',
    3: 'teal'
}

df['colores'] = pred

fig, ax = plt.subplots(figsize=(12,7))
ax.scatter(df['X'],df['Y'],c=df['colores'].map(colores),edgecolors='black')
plt.show()
```

```python
num = 1000

numeros = np.random.rand(num)
int_x = np.random.randint(-200,600,size=num)
int_y = np.random.randint(-200,800,size=num)
eje_x = np.sum([int_x, numeros], axis=0)
eje_y = np.sum([int_y, numeros], axis=0)
puntos = []

for i in zip(eje_x, eje_y):
    puntos.append(i)

df2 = pd.DataFrame(puntos, columns=['X','Y'])

```

```python
X2 = df2.to_numpy()
pred2 = modelo.predict(X2)

df2['colores'] = pred2

fig = plt.figure(figsize=(12,7))
ax = plt.subplot(1,1,1)
plt.scatter(df2['X'],df2['Y'], c=df2['colores'].map(colores), edgecolors='black')
plt.show()
```

```python
num = 600
cumulo1 = []
X = np.linspace(0,100,num)
Y = np.linspace(0,100,num)

Xmod = X * np.random.randn(num) + 250
Ymod = Y * np.random.randn(num) + 270

for i in zip(Xmod, Ymod):
    cumulo1.append(i)

df3 = pd.DataFrame(cumulo1, columns=['X','Y'])
```

```python
X3 = df3.to_numpy()
pred3 = modelo.predict(X3)
df3['colores'] = pred3
```
