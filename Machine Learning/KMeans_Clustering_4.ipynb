```python
import pandas as pd
from Generadores import *
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
datos = DatosSVM23D(400)

df = pd.DataFrame(datos, columns=['X','Y','Z'])

fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection='3d')
ax.scatter(df['X'],df['Y'],df['Z'],edgecolors='black')
plt.show()
```

```python
X = df.to_numpy()

modelo = KMeans(n_clusters=8)
modelo.fit(X)
pred = modelo.predict(X)
```

```python
colores = {
    0: 'crimson',
    1: 'orchid',
    2: 'mediumpurple',
    3: 'teal',
    4: 'magenta',
    5: 'sienna',
    6: 'gold',
    7: 'cyan'
}

df['colores'] = pred

fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection='3d')
ax.scatter3D(df['X'],df['Y'],df['Z'],c=df['colores'].map(colores),edgecolors='black')
plt.show()
```

```python
num = 1000
lista = []

numeros = np.random.rand(num)

int_x = np.random.randint(-200,800,size=num)
int_y = np.random.randint(-200,1000,size=num)
int_z = np.random.randint(-200,600,size=num)

eje_x = np.sum([int_x, numeros], axis=0)
eje_y = np.sum([int_y, numeros], axis=0)
eje_z = np.sum([int_z, numeros], axis=0)

for i in zip(eje_x, eje_y, eje_z):
    lista.append(i)

df2 = pd.DataFrame(lista, columns=['X','Y','Z'])
```

```python
X2 = df2.to_numpy()
pred2 = modelo.predict(X2)
```

```python
df2['colores'] = pred2

fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection='3d')
ax.scatter3D(df2['X'],df2['Y'],df2['Z'],c=df2['colores'].map(colores),edgecolors='black')
```
