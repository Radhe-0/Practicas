```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Generadores import *
```

```python
datos = DatosSVM(150)
df = pd.DataFrame(datos, columns=['X','Y'])
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(df['X'],df['Y'], c='crimson',edgecolors='black')
```

```python
X = df['X'].to_numpy().reshape(-1,1)
Y = df['Y'].to_numpy()
```

```python
modelo = KMeans(n_clusters=2)
modelo.fit(X)
```

```python
predicciones = modelo.predict(X)
```

```python
colors = {0: 'slateblue', 
          1: 'teal'}

fig, ax = plt.subplots(figsize=(12,8))
df['color'] = predicciones
ax.scatter(df['X'],df['Y'],c=df['color'].map(colors),edgecolors='black')
```
