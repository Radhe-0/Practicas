## Ventas anuales de varias empresas

```python
periodo = np.arange(2010,2020,1)

ventas_empresa1 = np.random.randint(50000,100000,size=len(periodo))
ventas_empresa2 = np.random.randint(70000,150000,size=len(periodo))
ventas_empresa3 = np.random.randint(90000,180000,size=len(periodo))
ventas_empresa4 = np.random.randint(110000,230000,size=len(periodo))

plt.grid()
plt.plot(periodo, ventas_empresa1, label='Empresa 1', marker='*',c='royalblue',linestyle='dashed')
plt.plot(periodo, ventas_empresa2, label='Empresa 2', marker='P',c='mediumslateblue',linestyle='dashed')
plt.plot(periodo, ventas_empresa3, label='Empresa 3', marker='v',c='seagreen',linestyle='dashed')
plt.plot(periodo, ventas_empresa4, label='Empresa 4', marker='o',c='indianred',linestyle='dashed')
plt.xlabel('Periodo')
plt.ylabel('Ventas')
plt.legend()
plt.minorticks_on()
plt.title('Ventas anuales')
plt.show()
```

```python
plt.bar(periodo,ventas_empresa4,color='indianred')
plt.bar(periodo,ventas_empresa3,color='seagreen',bottom=ventas_empresa4)
plt.bar(periodo,ventas_empresa2,color='mediumslateblue',bottom=ventas_empresa3+ventas_empresa4)
plt.bar(periodo,ventas_empresa1,color='royalblue',bottom=ventas_empresa2+ventas_empresa3+ventas_empresa4)
plt.xlabel('Periodo')
plt.ylabel('Ventas')
plt.minorticks_on()
plt.title('Ventas anuales')
plt.show()
```

```python
plt.bar(periodo,ventas_empresa4,color='indianred',width=0.2)
plt.bar(periodo+0.2,ventas_empresa3,color='seagreen',width=0.2)
plt.bar(periodo+0.4,ventas_empresa2,color='mediumslateblue',width=0.2)
plt.bar(periodo+0.6,ventas_empresa1,color='royalblue',width=0.2)
plt.xlabel('Periodo')
plt.ylabel('Ventas')
plt.minorticks_on()
plt.title('Ventas anuales')
plt.show()
```

```python
plt.barh(periodo,ventas_empresa4,color='indianred',height=0.2)
plt.barh(periodo+0.2,ventas_empresa3,color='seagreen',height=0.2)
plt.barh(periodo+0.4,ventas_empresa2,color='mediumslateblue',height=0.2)
plt.barh(periodo+0.6,ventas_empresa1,color='royalblue',height=0.2)
plt.xlabel('Ventas')
plt.ylabel('Periodo')
plt.minorticks_on()
plt.title('Ventas anuales')
plt.show()
```

```python
plt.bar(periodo, ventas_empresa1, width=1/3,color='royalblue')
plt.bar(periodo+0.3333, ventas_empresa2, width=1/3,color='mediumslateblue')
plt.plot(periodo,ventas_empresa3,c='seagreen')
plt.plot(periodo,ventas_empresa4,c='indianred')
plt.xlabel('Periodo')
plt.ylabel('Ventas')
plt.minorticks_on()
plt.title('Ventas anuales')
plt.show()
```
