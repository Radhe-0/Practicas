```python
from Generadores import * # Importo el módulo para poder generar datos
import matplotlib.pyplot as plt # Importo matplotlib

datos = DatosAltura(350) # Genero los datos
df = pd.DataFrame(datos, columns=['Personas','Alturas']) # Los cargo en un dataframe

# Grafico los datos
plt.grid()
plt.scatter(x='Alturas',y='Personas',data=df,edgecolors='navy',c='plum', marker='^',linewidths=1.2)
plt.ylabel('Alturas')
plt.xlabel('Personas')
plt.xticks(np.arange(0,1.1,0.1))
plt.yticks(np.arange(1.3,2.3,0.1))
plt.title('Alturas de un conjunto de personas')
plt.minorticks_on( )
plt.show()
```

```python
periodo = np.arange(2000,2020,1)

ventas_empresa1 = np.random.randint(50000,100000,size=len(periodo))
ventas_empresa2 = np.random.randint(70000,150000,size=len(periodo))
ventas_empresa3 = np.random.randint(90000,180000,size=len(periodo))
ventas_empresa4 = np.random.randint(110000,230000,size=len(periodo))

plt.grid()
plt.plot(periodo, ventas_empresa1, label='Empresa 1', marker='*',c='teal')
plt.plot(periodo, ventas_empresa2, label='Empresa 2', marker='P',c='mediumslateblue')
plt.plot(periodo, ventas_empresa3, label='Empresa 3', marker='v',c='olive')
plt.plot(periodo, ventas_empresa4, label='Empresa 4', marker='o',c='indianred')
plt.legend()
plt.minorticks_on()
plt.xticks(np.linspace(2000,2020,5))
plt.title('Ventas anuales')
plt.show()
```
